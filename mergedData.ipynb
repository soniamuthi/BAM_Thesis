```python
import pandas as pd
```

## Merge the 4 datasets and do pre-processing where required

```python
data2019 = pd.read_csv("2019data.csv")
data2020 = pd.read_csv("2020data.csv")
data2021 = pd.read_csv("2021data.csv")
data2022 = pd.read_csv("2022data.csv")
```

```python
data2019 = data2019.drop("PrevPodiumRace", axis = 1)
```

```python
final_df = pd.concat([data2019, data2020, data2021, data2022], axis=0, ignore_index=True)
```

```python
final_df = final_df.drop("Unnamed: 0", axis =1)
```

```python
a = final_df[final_df["Abbreviation"] == "GIO"]
duplicate_rows = a.duplicated()
num_duplicates = duplicate_rows.sum()
print("Number of duplicate rows:", num_duplicates)
```

```python
# Remove duplicates
final_df = final_df.drop_duplicates()
final_df
```

```python
# Check for NaN values
nan_df = final_df.isna()

num_true_values = nan_df.sum().sum()

print(num_true_values)
```

```python
## Add race type variable
countries = ['Monaco', 'Singapore', 'Australia', 'Canada', 'Russia']
final_df['CircuitType'] = final_df['RaceCountry'].apply(lambda x: 'street' if x in countries else 'race')
```

# Transformation of variables type

```python
final_df["AvgPitTime"].dtypes
```

```python
def str_to_seconds(time_str):
    time_td = pd.to_timedelta(str(time_str))
    return time_td.total_seconds()

# apply the function to the timedelta columns
timedelta_cols = ['BestQualiTime', 'AvgLapTime', 'SDLapTime', 'AvgSplitTime', 'AvgPitTime']
for col in timedelta_cols:
    final_df[col] = final_df[col].apply(str_to_seconds)
```

```python
final_df
```

```python
# Replace values for Status variable
final_df['Status'].unique()
```

```python
# replace car issues
car_issues = ['Damage', 'Engine', 'Wheel', 'Out of fuel', 'Collision damage', 'Collision', 'Power Unit', 'Transmission', 'Brakes', 'Suspension', 'Spun off', 'Exhaust', 'Power loss', 'Water pressure', 'Hydraulics', 'Disqualified', 'Overheating', 'Oil leak', 'Electronics', 'Fuel pressure', 'Debris', 'Puncture', 'Radiator', 'Gearbox', 'Electrical', 'Wheel nut', 'Turbo', 'Rear wing', 'Mechanical', 'Water pump', 'Cooling system', 'Fuel leak', 'Front wing', 'Water leak', 'Vibrations', 'Fuel pump', 'Undertray', 'Differential']
final_df['Status'] = final_df['Status'].replace(car_issues, 'carIssue')

# replace driver issues
driver_issues = ['Illness', 'Retired', 'Accident', 'Withdrew']
final_df['Status'] = final_df['Status'].replace(driver_issues, 'driverIssue')

# replace late finish
late_finish = ['+1 Lap', '+2 Laps', '+3 Laps', '+5 Laps', '+6 Laps']
final_df['Status'] = final_df['Status'].replace(late_finish, 'lateFinished')
```

```python
final_df = final_df.rename(columns={'FLapPoint': 'FLap'})
```

```python
# Remove fastest lap point
final_df['Points'] = final_df['Points'].apply(lambda x: x-1 if x in [26, 19, 16, 13, 11, 9, 7, 5, 3] else x)
```

```python
# Check if there is the case for fastest lap for 1
a = final_df[["Points", "FLap"]]
a[(a["FLap"] == 1) & (a["Points"] == 1)]
```

```python
final_df.loc[(final_df['TeamName'] == 'AlphaTauri') & (final_df['TeamName'].isnull()), 'TeamName'] = 'Honda'
final_df.loc[(final_df['TeamName'] == 'Alfa Romeo Racing') & (final_df['TeamName'].isnull()), 'TeamName'] = 'Ferrari'
```

```python
# Replace all the observations were grid position is 0 to 20
final_df['GridPosition'] = final_df['GridPosition'].replace(0, 20)
```

```python
# Save the final dataset
final_df.to_csv(r'mergedData.csv', index=True, header=True)
```
